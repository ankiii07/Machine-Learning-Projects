## Import Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Iris.csv')
```

```python
df.head()
```

```python
df.shape
```

### Univariate Analysis

```python
df_setosa = df.loc[df['Species']=='Iris-setosa']
df_virginica = df.loc[df['Species']=='Iris-virginica']
df_versicolor = df.loc[df['Species']=='Iris-versicolor']
```

```python
plt.plot(df_setosa['SepalLengthCm'],np.zeros_like(df_setosa['SepalLengthCm']),'*')
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.plot(df_versicolor['SepalLengthCm'],np.zeros_like(df_versicolor['SepalLengthCm']),'+')
```

```python
plt.plot(df_setosa['SepalWidthCm'],np.zeros_like(df_setosa['SepalWidthCm']),'*')
plt.xlabel('SepalLength')
plt.show()
```

```python
plt.plot(df_virginica['SepalLengthCm'],np.zeros_like(df_virginica['SepalLengthCm']),'o')
plt.xlabel('SepalLength')
plt.show()
```

```python
plt.plot(df_versicolor['PetalLengthCm'],np.zeros_like(df_versicolor['PetalLengthCm']),'+')
plt.xlabel('PetalLengthCm')
plt.show()
```

### Bivariate Analysis

```python
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,"SepalLengthCm","SepalWidthCm").add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,"PetalLengthCm","PetalWidthCm").add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend();
plt.show()
```

```python
sns.FacetGrid(df,hue='Species',size=5).map(plt.scatter,"SepalLengthCm","PetalWidthCm").add_legend();
plt.show()
```

### Multivariate Analysis

```python
sns.histplot(data=df, x="SepalWidthCm", hue="Species")
```

```python
sns.histplot(data=df, x="SepalWidthCm", hue="Species",multiple="stack")
```

```python
sns.histplot(data=df, x="SepalLengthCm", hue="Species",element="step")
```

```python
sns.histplot(data=df, x="PetalLengthCm", hue="Species",element="poly")
```

```python
sns.histplot(
    df, x="SepalLengthCm", hue="Species", element="step",
    stat="density", common_norm=False,
)
```

```python
sns.histplot(data=df, x="SepalLengthCm",log_scale=True, element="step", fill=False)
```

```python
sns.histplot(
    data=df, x="SepalWidthCm", hue="Species",
    hue_order=["Iris-Versicolor", "Iris-virginica"],
    log_scale=True, element="step", fill=False,
    cumulative=True, stat="density", common_norm=False,
)
```

```python
sns.histplot(
    df, x="PetalLengthCm", y="Species", hue="Species", legend=False
)
```
